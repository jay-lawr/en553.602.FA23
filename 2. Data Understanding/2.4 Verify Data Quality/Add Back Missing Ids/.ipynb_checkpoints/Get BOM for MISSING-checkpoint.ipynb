```python
import boxofficemojoscrape as box
import pandas as pd
```

```python
df = pd.read_csv("MISSING_imdb_ids.csv")
```

```python
results = []
imdb_ids_that_failed = []
counter = 1
for id in df.imdb_id:
    print(counter)
    try:
        data = box.get_data_by_id(id)
        results.append(data)
    except:
        imdb_ids_that_failed.append(id)
    counter += 1

results_df = pd.DataFrame(results)
results_df.to_csv("BOM_new_data.csv")

failed_df = pd.DataFrame(imdb_ids_that_failed)
failed_df.to_csv("MISSING_Failed_IDs.csv")
```
