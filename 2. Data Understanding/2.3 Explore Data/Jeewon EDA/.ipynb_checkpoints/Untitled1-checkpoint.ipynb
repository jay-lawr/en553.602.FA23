```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataset is stored in a pandas DataFrame called 'df'
# Replace 'revenue_column' with the actual column name for revenue in your dataset
# Replace 'mpaa_column' with the actual column name for MPAA rating in your dataset

# Example:
df = pd.read_csv('../../../3. Data Preparation/3.3 Construct Data/merged_data_mod_scores_filtered_US_release_year_star_scores_prod_scores_dir_scores.csv')
revenue_column = 'IMDB Domestic Revenue'
mpaa_column = 'MPAA'

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Create a box plot
sns.boxplot(x=mpaa_column, y=revenue_column, data=df, ax=ax)

# Set plot labels and title
ax.set(xlabel='MPAA Rating', ylabel='Revenue', title='MPAA Rating vs. Revenue')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your dataset is stored in a pandas DataFrame called 'df'
# Replace 'unweighted_star_score_column' with the actual column name for unweighted star scores in your dataset

# Example:
unweighted_star_score_column = 'Total Production Company Score'

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Create a KDE plot for unweighted star scores
sns.kdeplot(data=df[unweighted_star_score_column], fill=True, ax=ax)

# Set plot labels and title
ax.set(xlabel='Total Production Company Score', ylabel='Density', title='Kernel Density Estimate for Unweighted Star Score')

# Show the plot
plt.show()
```
