```python
import pandas as pd
```

```python
df = pd.read_csv("../../3. Data Preparation/3.3 Construct Data/merged_data_mod_scores.csv")
df = df.dropna(subset = ["IMDB Domestic Revenue"])
```

```python
for col in df.columns:
    print(col)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset

# Display basic summary statistics
print(df.describe())


# Scatter plot of IMDB Budget vs. Domestic Revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TMDB Budget', y='IMDB Domestic Revenue', data=df)
plt.title('TMDB Budget vs. IMDB Domestic Revenue')
plt.xlabel('TMDB Budget')
plt.ylabel('IMDB Domestic Revenue')
plt.show()

# Box plot of Domestic Distributor and Domestic Revenue
plt.figure(figsize=(14, 8))
sns.boxplot(x='Domestic Distributor', y='IMDB Domestic Revenue', data=df)
plt.title('Domestic Distributor vs. IMDB Domestic Revenue')
plt.xlabel('Domestic Distributor')
plt.ylabel('IMDB Domestic Revenue')
plt.xticks(rotation=90)
plt.show()
```
