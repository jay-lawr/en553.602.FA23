```python
import pandas as pd
```

```python
df = pd.read_csv("BOM_full_data.csv")
ids = pd.read_csv("imdb_ids.csv")
failed_ids = pd.read_csv("Failed_IDs.csv")
```

```python
list(failed_ids['0'])
```

```python
correct_ids = []
for id in list(ids.imdb_id):
    if id not in list(failed_ids['0']):
        correct_ids.append(id)
print(len(correct_ids), len(df))
```

```python
df["imdb_ids"] = correct_ids
```

```python
df.to_csv("BOM_full_data.csv")
```
