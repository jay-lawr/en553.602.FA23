```python
import pandas as pd

filename='merged_US_allscores_similarity.csv'
df = pd.read_csv(filename)
col = df['Domestic Distributor']
col.fillna('Empty', inplace=True)
distributor_ids, unique_labels = pd.factorize(col)
df['Domestic Distributor ID'] = distributor_ids

# Print the DataFrame to see the changes
print(df.head)
```

```python
df.to_csv(filename, index=False)
```
