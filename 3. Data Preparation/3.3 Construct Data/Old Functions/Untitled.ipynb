```python
# Initialize csv file paths
csv_file_path ='merged_data_mod_scores.csv'
# Create DataFrames from both CSV files
df = pd.read_csv(csv_file_path)
```

```python

bins = [0, 1000000, 5, 10, 25, 50, 100]
df['binned'] = pd.cut(df['percentage'], bins)
df[['Unweighted Star Score']].hist(bins=10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the revenue brackets (adjust as needed)
revenue_bins = [100000, 1000000, 10000000, 50000000, 100000000, 500000000, float('inf')]
revenue_labels = ["100000-1000000", "1000000-10000000", "10000000-50000000", "50000000-100000000", "100000000-500000000", "500000000+"]
df['revenue_bracket'] = pd.cut(df['Merged Revenue'], bins=revenue_bins, labels=revenue_labels, right=False)

# Create a box plot of "star power" scores in different revenue brackets
plt.figure(figsize=(10, 6))
plt.xticks(rotation=45)
plt.xlabel('Revenue Bracket')
plt.ylabel('Star Power')
plt.title('Box Plot of Star Power Scores in Different Revenue Brackets')

plt.boxplot([df[df['revenue_bracket'] == label]['Unweighted Star Score'] for label in revenue_labels], labels=revenue_labels)

plt.show()
```
